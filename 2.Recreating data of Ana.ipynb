```python
import pandas as pd
import matplotlib.pyplot as plt

Masses = [85, 90, 100, 110, 130]
colors = ['r', 'g', 'b', 'c', 'y']  # Colors for each mass

# Initialize empty lists to store the data
PDT_list = []
Z_list = []

# Iterate over the masses
for mass in Masses:
    PDT = []
    Z = []
    
    # Iterate over the files
    for file in file_list:
        df = pd.read_hdf(file, key=None) 
        
        if df['A'][0] == mass:
            PDT.append(df['PDT'][0])
            Z.append(df['E'][0])
    
    # Rearrange the lists based on Z values in ascending order
    PDT, Z = zip(*sorted(zip(PDT, Z), key=lambda x: x[1]))
    
    # Store the data in the lists
    PDT_list.append(PDT)
    Z_list.append(Z)

# Plot the data for each mass
for i in range(len(Masses)):
    plt.plot(Z_list[i], PDT_list[i], color=colors[i], marker='o', linestyle='', label=f'A = {Masses[i]}')
    plt.plot(Z_list[i], PDT_list[i], color=colors[i], linestyle='-', label='', alpha=0.5)

plt.title('PDT calculations for different masses')
plt.ylabel('Time [ns]')
plt.xlabel('Energy [MeV]')
plt.legend()
plt.grid()
plt.show()
```
